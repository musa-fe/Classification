Veride birbirini tekrar eden sÃ¼tunlar var. KorelasyonlarÄ± 0.90 olan sÃ¼tunlardan bazÄ±larÄ±nÄ± veriden atÄ±p sadeleÅŸtirelim. Bu kodda Ã¼st Ã¼Ã§gen kullanÄ±larak korelasyonlarÄ± yÃ¼ksek olan iki sÃ¼tundan sadece birini veriden siliyoruz. Tabi rastgele bir silme yapmayacaÄŸÄ±z hedef deÄŸiÅŸken ile daha fazla iliÅŸkili olan sÃ¼tunu tutup diÄŸerini sileceÄŸiz

```python
df['diagnosis_numeric'] = df['diagnosis'].map({'M': 1, 'B': 0})

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

corr_matrix = df[numerical_cols].corr().abs()

upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

high_corr_pairs = [(col1, col2)
                   for col1 in upper.columns
                   for col2 in upper.index
                   if upper.loc[col2, col1] > 0.9]

target_corr = df[numerical_cols].corrwith(df['diagnosis_numeric']).abs()

columns_to_drop = []
for col1, col2 in high_corr_pairs:
    if target_corr[col1] >= target_corr[col2]:
        columns_to_drop.append(col2)
    else:
        columns_to_drop.append(col1)

columns_to_drop = list(set(columns_to_drop))

df_cleaned = df.drop(columns=columns_to_drop)

print("Silinen sÃ¼tunlar:", columns_to_drop)
print("Yeni shape:", df_cleaned.shape)
```

VaryanslarÄ± Ã§ok dÃ¼ÅŸÃ¼k olan sÃ¼tunlara bakalÄ±m varsa onlarÄ±da silebiliriz

texture_se de dÃ¼ÅŸÃ¼k bir korelasyona sahip onu da atalÄ±m veriden. GÃ¶zden kaÃ§mÄ±ÅŸ olabilir

```python
df_model = df_model.drop(columns=['texture_se'])
df_model.corrwith(df_model['diagnosis_numeric']).sort_values(ascending=False)
```

```python
features = df_model.drop(columns=['diagnosis_numeric']).columns

outlier_summary = {}

for col in features:
    Q1 = df_model[col].quantile(0.25)
    Q3 = df_model[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df_model[(df_model[col] < lower_bound) | (df_model[col] > upper_bound)]

    outlier_summary[col] = len(outliers)

sorted_outliers = sorted(outlier_summary.items(), key=lambda x: x[1], reverse=True)
for col, count in sorted_outliers[:12]:
    print(f"{col}: {count} aykÄ±rÄ± deÄŸer")
```

clip kullanarak verimizdeki aykÄ±rÄ± deÄŸerleri yok ettik. Model kurarken aykÄ±rÄ± deÄŸerlere duyarlÄ± olan modellerde bu temiz veriyi kullanÄ±rÄ±z.

*   df_model  = ( aykÄ±rÄ± deÄŸerli verimiz )
*   df_model1 = ( aykÄ±rÄ± deÄŸerlerden arÄ±ndÄ±rÄ±ÅŸmÄ±ÅŸ verimiz)



```python
df_model1 = df_model.copy()

features = df_model1.drop(columns=['diagnosis_numeric']).columns

for col in features:
    Q1 = df_model1[col].quantile(0.25)
    Q3 = df_model1[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_model1[col] = df_model1[col].clip(lower=lower_bound, upper=upper_bound)

outlier_summary = {}

for col in features:
    Q1 = df_model1[col].quantile(0.25)
    Q3 = df_model1[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df_model1[(df_model1[col] < lower_bound) | (df_model1[col] > upper_bound)]

    outlier_summary[col] = len(outliers)

sorted_outliers = sorted(outlier_summary.items(), key=lambda x: x[1], reverse=True)
for col, count in sorted_outliers[:12]:
    print(f"{col}: {count} aykÄ±rÄ± deÄŸer")
```

MODEL KURMAYA GEÃ‡EBÄ°LÄ°RÄ°Z

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

X1 = df_model1.drop(columns=['diagnosis_numeric'])
y1 = df_model1['diagnosis_numeric']

pipeline1 = make_pipeline(StandardScaler(), LogisticRegression(max_iter=1000))

cv_acc = cross_val_score(pipeline1, X1, y1, cv=5, scoring='accuracy')
cv_f1 = cross_val_score(pipeline1, X1, y1, cv=5, scoring='f1')

print("DF_MODEL1 Logistic Regression")
print("- Ortalama Accuracy:", np.mean(cv_acc))
print("- Ortalama F1:", np.mean(cv_f1))
```
